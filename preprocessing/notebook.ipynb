```python
import pandas as pd

df = pd.read_csv('../../data/raw/mra_train.csv')
df.head()
```

```python
liste = df['SeriesInstanceUID'].tolist()
```

```python
import os

mapping_dir = '../../data/preprocessed_data/mapping'

# Get all JSON files in the mapping directory
all_files = [f for f in os.listdir(mapping_dir) if f.endswith('_mapping.json')]

# Extract series IDs from filenames
existing_series_ids = [f.replace('_mapping.json', '') for f in all_files]

# Find series IDs to delete (those not in liste)
to_delete = [sid for sid in existing_series_ids if sid not in liste]

len([sid for sid in existing_series_ids if sid in liste])
```

```python
# Delete files not in the liste
for series_id in to_delete:
    file_path = os.path.join(mapping_dir, f'{series_id}_mapping.json')
    if os.path.exists(file_path):
        os.remove(file_path)
        print(f'Deleted: {file_path}')
```
